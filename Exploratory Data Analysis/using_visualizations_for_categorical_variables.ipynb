```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Iris dataset
iris = sns.load_dataset("iris")
```

```python
iris.head()
# iris.info()
# iris.describe()
```

```python
# Quantity species
iris['species'].value_counts()
```

```python
col = 'sepal_length'
sns.histplot(data=iris, x='sepal_length',kde=True).set_title("Distribuition of variable sepal_lenght")
```

```python
for col in iris.drop(columns='species'):
    sns.histplot(data=iris, x=col, kde=True).set_title(f"{col}")
    plt.show()
```

```python
for col in iris.drop(columns='species'):
    sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(f"{col}")
    plt.show()
```

```python
for col in iris.drop(columns='species'):
    sns.barplot(data=iris, x=iris['species'], hue=iris['species'],y=col, errorbar=('ci', 90))
    plt.show()
```
