```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
candy_df = pd.read_csv("datasets/candy_production.csv")
```

```python
# candy_df.head()
# test_df.tail()
candy_df.head(15)
```

```python
candy_df.info()
```

```python
# Cast for date
candy_df['observation_date'] = pd.to_datetime(candy_df["observation_date"])
```

```python
# Creation Graphical with no filtered
candy_df.plot(x="observation_date", y="IPG3113N", kind='line', label="historical series",grid=True)
```

```python
#Create base with filtered
candy_df_filtered = candy_df[candy_df['observation_date'] >= '2011-01-01'] 
```

```python
candy_df_filtered.tail()
```

```python
candy_df_filtered.plot(x="observation_date", y="IPG3113N",grid=True, kind="line", label="historical series", figsize=(12, 4))
```

```python
ax = candy_df_filtered.plot(x="observation_date", y= "IPG3113N", kind="line", figsize= (12, 4), label="historical serie")
x_cords = ["2011-01-01","2012-01-01","2013-01-01", "2014-01-01", "2015-01-01", "2016-01-01", "2017-01-01"]
```

```python
ax = candy_df_filtered.plot(x="observation_date", y= "IPG3113N", kind="line", figsize= (12, 4), label="historical serie")

x_cords = ["2011-01-01","2012-01-01","2013-01-01", "2014-01-01", "2015-01-01", "2016-01-01", "2017-01-01"]

for xc in x_cords:
    plt.axvline(x = xc, color="black", linestyle='--')
```
